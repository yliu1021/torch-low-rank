```python
from glob import glob
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
plt.ioff()
```

```python
results = [json.load(open(f)) for f in glob("set_rank_results_cifar10/*.json")]
results = pd.DataFrame(results)
def get_epoch(x):
    return np.argmax(x) + 1

results["best_loss"] = results["loss"].map(np.min)
results["best_loss_epoch"] = results["loss"].map(get_epoch)
results["best_categorical_accuracy"] = results["categorical_accuracy"].map(np.max)
results["best_categorical_accuracy_epoch"] = results["categorical_accuracy"].map(get_epoch)
results["best_val_loss"] = results["val_loss"].map(np.min)
results["best_val_loss_epoch"] = results["val_loss"].map(get_epoch)
results["best_val_categorical_accuracy"] = results["val_categorical_accuracy"].map(np.max)
results["best_val_categorical_accuracy_epoch"] = results["val_categorical_accuracy"].map(get_epoch)
results["initial_rank"].replace(-1, 256, inplace=True)
results["new_rank"].replace(-1, 256, inplace=True)
results.head()
```

```python
def plottable_3d_info(df: pd.DataFrame):
    index = df.index
    columns = df.columns
    x, y = np.meshgrid(np.arange(len(columns)), np.arange(len(index)))
    z = np.array([[df[c][i] for c in columns] for i in index])
    x_ticks = dict(ticks=np.arange(len(columns)), labels=columns)
    y_ticks = dict(ticks=np.arange(len(index)), labels=index)
    return x, y, z, x_ticks, y_ticks
```

```python
def wireframe_plot(r, z_str):
    x, y, z, x_ticks, y_ticks = plottable_3d_info(r)
    fig = plt.figure()
    axes = fig.add_subplot(projection="3d")
    axes.plot_wireframe(x, y, z)
    axes.set_xlabel('New Rank')
    axes.set_ylabel('Rank Update Epoch')
    axes.set_zlabel(z_str)
    axes.set_zlim3d(bottom=0)
    plt.xticks(**x_ticks)
    plt.yticks(**y_ticks)
    plt.show()

def rank_update_analysis(initial_rank: int):
    r = pd.pivot(
        results[results["initial_rank"] == initial_rank],
        index="rank_update_epoch",
        columns="new_rank")
    wireframe_plot(r["best_val_loss"], "Best Val Loss")
    wireframe_plot(r["best_val_loss_epoch"], "Best Val Epoch")

rank_update_analysis(160)
```

```python
def rank_update_epoch_analysis(initial_rank: int, new_rank: int):
    r = results.loc[(results["initial_rank"] == initial_rank) & (results["new_rank"] == new_rank)]
    r.groupby("rank_update_epoch").first()  # we should only have one run for each config

    plt.xlabel("Rank Update Epoch")
    plt.ylabel("Val Loss")
    plt.plot(r["rank_update_epoch"], r["best_val_loss"])
    plt.show()

    plt.xlabel("Rank Update Epoch")
    plt.ylabel("Best Val Loss Epoch")
    plt.plot(r["rank_update_epoch"], r["best_val_loss_epoch"])
    plt.show()
rank_update_epoch_analysis(160, 10)
```

```python
def new_rank_analysis(initial_rank: int, rank_update_epoch: int):
    r = results.loc[(results["initial_rank"] == initial_rank) & (results["rank_update_epoch"] == rank_update_epoch)]
    r.groupby("rank_update_epoch").first()  # we should only have one run for each config

    plt.xlabel("New Rank")
    plt.ylabel("Val Loss")
    plt.plot(r["new_rank"], r["best_val_loss"])
    plt.show()

    plt.xlabel("New Rank")
    plt.ylabel("Best Val Loss Epoch")
    plt.plot(r["new_rank"], r["best_val_loss_epoch"])
    plt.show()
new_rank_analysis(160, 1)
```
